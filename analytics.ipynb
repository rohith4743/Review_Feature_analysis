```python
#  lexical diversity
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Preprocessing function
def preprocess_text(text):
    # Tokenize
    tokens = word_tokenize(text.lower())
    
    # Remove punctuation and non-alphabetic characters
    tokens = [token for token in tokens if token.isalpha()]
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    
    return tokens


def lexical_diversity(text):
    tokens = preprocess_text(text)
    unique_tokens = set(tokens)
    return len(unique_tokens) / len(tokens) if tokens else 0

# Applying lexical diversity calculation
df['lexical_diversity'] = df['text_'].progress_apply(lexical_diversity)

df.head()
```

```python
import spacy
from spellchecker import SpellChecker
nlp = spacy.load("en_core_web_sm")

# Initialize spell checker
spell = SpellChecker()

# Function to perform NER and spell check
def ner_spellcheck(text):
    # Process the text with spaCy
    doc = nlp(text)
    # Extract words that are not part of named entities
    non_entity_words = [token.text for token in doc if not token.ent_type_]
    # Check for misspellings in the non-entity words
    misspelled = spell.unknown(non_entity_words)
    return misspelled

df['misspelled'] = df['text_'].progress_apply(ner_spellcheck)
df.head()
```

```python
import tensorflow as tf
from transformers import BertTokenizer, TFBertModel
from tqdm.auto import tqdm
import pandas as pd
import numpy as np
import gc 
import json
import os

df = pd.read_csv("analysed.csv")
# Setup
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = TFBertModel.from_pretrained('bert-base-uncased')

def compute_coherence_scores(embeddings):
    coherence_scores = []
    for i in range(len(embeddings) - 1):
        cos = tf.keras.losses.cosine_similarity(embeddings[i:i+1], embeddings[i+1:i+2])
        coherence_scores.append(cos.numpy().squeeze())
    return coherence_scores


def process_batch(texts):
    inputs = tokenizer(texts, return_tensors="tf", padding=True, truncation=True, max_length=512)
    outputs = model(inputs)
    embeddings = tf.reduce_mean(outputs.last_hidden_state, axis=1)
    return embeddings

def batch_narrative_coherence(batch_texts):
    # Split texts into sentences and filter out any empty texts
    batch_sentences = [[x.strip() for x in text.split(".") if x] for text in batch_texts]
    flat_texts = [sentence for sublist in batch_sentences for sentence in sublist]

    # Process embeddings in a batch
    if flat_texts:
        embeddings = process_batch(flat_texts)
        # Reconstruct the batch from flat list of embeddings
        start = 0
        coherence_scores = []
        for sentences in batch_sentences:
            num_sentences = len(sentences)
            if num_sentences > 1:
                sentence_embeddings = embeddings[start:start+num_sentences]
                scores = compute_coherence_scores(sentence_embeddings)
                coherence_scores.append(np.mean(scores))
            else:
                coherence_scores.append(0.0)
            start += num_sentences
        return coherence_scores
    else:
        return [0.0] * len(batch_texts)

def process_chunk(df, chunk_size, chunk_number, json_file):
    
    
    start_idx = chunk_number * chunk_size
    end_idx = start_idx + chunk_size
    if end_idx > len(df):
        end_idx = len(df)

    subset_df = df.iloc[start_idx:end_idx]

    batch_size = 32 
    results = []
    for i in tqdm(range(0, len(subset_df), batch_size), desc=f"Processing Chunk {chunk_number}"):
        batch_texts = subset_df['text_'][i:i+batch_size].tolist()
        batch_scores = batch_narrative_coherence(batch_texts)
        
        batch_scores = [float(score) for score in batch_scores]
        results.extend(batch_scores)
        
        tf.keras.backend.clear_session()
        gc.collect()

    if os.path.exists(json_file):
        with open(json_file, 'r') as file:
            json_data = json.load(file)
    else:
        json_data = {}

    json_data[chunk_number] = results

    with open(json_file, 'w') as file:
        json.dump(json_data, file, indent=4)

    return



chunk_number = 6 #range 1-14
chunk_size = 3200  
json_file = "narrative_coherence.json"

# process_chunk(df, chunk_size, chunk_number, json_file)

```

```python
import string
# Define a list of commonly used punctuations you want to track
common_punctuations = [',', '.', '?', ';', ':', '!']

# Function to count punctuation
def count_punctuations(text):
    # Count of all punctuations
    all_punct_count = sum(1 for char in text if char in string.punctuation)
    # Count of specific commonly used punctuations
    specific_punct_count = {punct: text.count(punct) for punct in common_punctuations}
    return all_punct_count, specific_punct_count

# Apply the function to each row in the DataFrame
df['punctuation_usage'] = df['text_'].progress_apply(count_punctuations)
df.head()
```

```python
# List of personal pronouns to track
import spacy
personal_pronouns = ['I', 'me', 'we', 'us', 'you', 'he', 'him', 'she', 'her', 'it', 'they', 'them']
nlp = spacy.load("en_core_web_sm")
# Function to count personal pronouns
def count_personal_pronouns(text):
    doc = nlp(text)
    pronoun_count = 0
    for token in doc:
        if token.text.lower() in personal_pronouns and token.pos_ == 'PRON':
            pronoun_count += 1
    return pronoun_count

# Apply the function to each row in the DataFrame
df['personal_pronoun_count'] = df['text_'].progress_apply(count_personal_pronouns)
df.head()
```

```python
import pandas as pd
import nltk
from nltk.util import ngrams
from nltk.corpus import stopwords
from collections import Counter
import string



# Function to preprocess text
def preprocess(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Tokenize text
    tokens = nltk.word_tokenize(text)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words and word.isalpha()]
    return tokens

# Function to generate n-grams
def generate_ngrams(tokens, n):
    return list(ngrams(tokens, n))

# Function to get top 10 n-grams for each label and n-gram size
def top_ngrams(df, n):
    # Group by 'label' and concatenate all texts into a single list per label
    grouped = df.groupby('label')['text_processed'].sum()
    # Calculate n-grams for each label
    top_ngrams = {}
    for label, tokens in grouped.items():
        # Generate n-grams
        all_ngrams = generate_ngrams(tokens, n)
        # Count and get top 10 n-grams
        ngram_counts = Counter(all_ngrams)
        top_ngrams[label] = ngram_counts.most_common(10)
    return top_ngrams

```

```python
# Preprocess texts
df['text_processed'] = df['text_'].apply(preprocess)

# Get top 10 n-grams for each n from 1 to 4
results = {}
for n in range(1, 5):
    results[f'Top 10 {n}-grams'] = top_ngrams(df, n)

# Display the results
for ngram_type, data in results.items():
    print(ngram_type)
    for label, ngrams in data.items():
        print(f"Label: {label}")
        for ngram, count in ngrams:
            print(f"{ngram}: {count}")
        print("\n")
```

```python
import tensorflow as tf
from transformers import BertTokenizer, TFBertModel
from tqdm.auto import tqdm
import pandas as pd
import numpy as np
import gc 
import json
import os

df = pd.read_csv("analysed.csv")
# Setup
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = TFBertModel.from_pretrained('bert-base-uncased')


def process_batch(texts):
    inputs = tokenizer(texts, return_tensors="tf", padding=True, truncation=True, max_length=512)
    outputs = model(inputs)
    embeddings = tf.reduce_mean(outputs.last_hidden_state, axis=1)
    return embeddings


def batch_contextual_relevance(reviews, categories):
    
    
    if reviews:
        
        review_embeddings = process_batch(reviews)
        category_embeddings = process_batch(categories)
      
        coherence_scores = []
        
        for review_emb, category_emb in zip(review_embeddings, category_embeddings):
            score = tf.keras.losses.cosine_similarity(review_emb, category_emb).numpy().squeeze()
            coherence_scores.append(score)
        return coherence_scores
    else:
        return []

def process_chunk(df, chunk_size, chunk_number, json_file):
    
    
    start_idx = chunk_number * chunk_size
    end_idx = start_idx + chunk_size
    if end_idx > len(df):
        end_idx = len(df)

    subset_df = df.iloc[start_idx:end_idx]

    batch_size = 32 
    results = []
    for i in tqdm(range(0, len(subset_df), batch_size), desc=f"Processing Chunk {chunk_number}"):
        batch_texts = subset_df['text_'][i:i+batch_size].tolist()
        batch_categories = subset_df['category'][i:i+batch_size].tolist()
        batch_scores = batch_contextual_relevance(batch_texts, batch_categories)
        
        batch_scores = [float(score) for score in batch_scores]
        results.extend(batch_scores)
        
        tf.keras.backend.clear_session()
        gc.collect()

    if os.path.exists(json_file):
        with open(json_file, 'r') as file:
            json_data = json.load(file)
    else:
        json_data = {}

    json_data[chunk_number] = results

    with open(json_file, 'w') as file:
        json.dump(json_data, file, indent=4)

    return



chunk_number = 6 
chunk_size = 9600  
json_file = "contextual_relevance.json"

# process_chunk(df, chunk_size, chunk_number, json_file)
```

```python
import tensorflow as tf
from transformers import BertTokenizer, TFBertForSequenceClassification
from tqdm.auto import tqdm
import pandas as pd
import numpy as np
import gc 
import json
import os

df = pd.read_csv("analysed.csv")

tokenizer = BertTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
model = TFBertForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment', num_labels=5)  # Assuming a 5-class model, adjust as needed



def process_batch_rating_relevance(texts):
    inputs = tokenizer(texts, return_tensors="tf", padding=True, truncation=True, max_length=512)
    outputs = model(inputs)
    logits = outputs.logits
    predictions = tf.nn.softmax(logits, axis=-1)
    return (tf.argmax(predictions, axis=-1) + 1).numpy().tolist()



def process_chunk(df, chunk_size, chunk_number, json_file):
    
    
    start_idx = chunk_number * chunk_size
    end_idx = start_idx + chunk_size
    if end_idx > len(df):
        end_idx = len(df)

    subset_df = df.iloc[start_idx:end_idx]

    batch_size = 128 
    results = []
    for i in tqdm(range(0, len(subset_df), batch_size), desc=f"Processing Chunk {chunk_number}"):
        batch_texts = subset_df['text_'][i:i+batch_size].tolist()
        batch_scores = process_batch_rating_relevance(batch_texts)
        
        batch_scores = [float(score) for score in batch_scores]
        results.extend(batch_scores)
        
        tf.keras.backend.clear_session()
        gc.collect()

    if os.path.exists(json_file):
        with open(json_file, 'r') as file:
            json_data = json.load(file)
    else:
        json_data = {}

    json_data[chunk_number] = results

    with open(json_file, 'w') as file:
        json.dump(json_data, file, indent=4)

    return



chunk_number = 4
chunk_size = 9600  
json_file = "rating_relevance.json"

# process_chunk(df, chunk_size, chunk_number, json_file)
```
